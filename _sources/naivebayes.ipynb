# Tugas Data Mining NaiveBayes

**Naive Bayes** adalah salah satu algoritma klasifikasi probabilistik yang sangat populer dalam data mining dan machine learning. Algoritma ini didasarkan pada teorema Bayes yang menyatakan bahwa probabilitas suatu hipotesis dapat dihitung berdasarkan probabilitas data yang diamati dan probabilitas a priori yang diketahui.

Naive Bayes disebut "naive" karena asumsi dasar yang digunakan oleh algoritma ini sangat sederhana dan tidak realistis, yaitu asumsi bahwa setiap atribut yang digunakan untuk mengklasifikasikan objek adalah independen satu sama lain. Meskipun asumsi ini seringkali tidak sepenuhnya benar dalam dunia nyata, algoritma Naive Bayes tetap memberikan hasil klasifikasi yang sangat baik dalam banyak kasus dan cukup efisien dalam komputasinya.

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/mk-gurucharan/Classification/master/IrisDataset.csv')
X = dataset.iloc[:,:4].values
y = dataset['species'].values
```

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/mk-gurucharan/Classification/master/IrisDataset.csv')
X = dataset.iloc[:,:4].values
y = dataset['species'].values
dataset.head(5)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)

```

```python
y_pred = classifier.predict(X_test) 
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
from sklearn.metrics import accuracy_score 
print ("Accuracy : ", accuracy_score(y_test, y_pred))
cm
```
